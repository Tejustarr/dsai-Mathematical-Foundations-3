# Univariate Derivative Rules — many examples & symbolic checks

```python
import sympy as sp
x=sp.symbols('x')
f=x**3*sp.sin(x)
sp.diff(f,x)
```

## Numeric verification using finite differences

```python
import numpy as np
f = lambda x: x**3 * np.sin(x)
def forward_diff(f,x,h=1e-6):
    return (f(x+h)-f(x))/h

a=0.7
print('symbolic=', float(sp.diff(f,x).subs(x,a)))
print('forward=', forward_diff(f,a,1e-6))
```

## Many derivative rules with examples: product, quotient, chain, log, exp

```python
g = sp.sin(sp.exp(x))
sp.diff(g,x)  # chain rule example
```
