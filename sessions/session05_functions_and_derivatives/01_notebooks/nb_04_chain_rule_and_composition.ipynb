# Chain rule for composition of functions — detailed matrix form and examples

```python
import sympy as sp
x,y = sp.symbols('x y')
# define F: R^2->R^2 and G: R^2->R
u,v = sp.symbols('u v')
F = sp.Matrix([x**2 + y, sp.sin(x*y)])
G = sp.Matrix([sp.exp(u+v)])
# Note: To demonstrate chain rule, substitute u=F1, v=F2
G_of_F = G.subs({u:F[0], v:F[1]})
# total derivative (Jacobian) via sympy
sp.Matrix([G_of_F[0]]).jacobian([x,y])
```
