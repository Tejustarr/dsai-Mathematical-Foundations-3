# Gram-Schmidt Process

```python
def gram_schmidt(V):
    U=[]
    for v in V.T:
        u=v.copy().astype(float)
        for q in U:
            u-=q*(q@v)/(q@q)
        U.append(u/np.linalg.norm(u))
    return np.array(U).T

V=np.array([[1,1],[1,0],[0,1]],float)
Q=gram_schmidt(V)
print(Q)
print('Q^TQ=\n', Q.T@Q)
```
